### Genero vs Cancelación

```python
def genero_vs_cancelacion(df):
    df_grouped = df.groupby(['customer.gender', 'Churn']).size().reset_index(name='Cantidad')

    # Porcentaje por género
    df_grouped['Total_por_genero'] = df_grouped.groupby('customer.gender')['Cantidad'].transform('sum')
    df_grouped['Porcentaje'] = round(df_grouped['Cantidad'] / df_grouped['Total_por_genero'] * 100, 1)

    # Crear texto combinado: cantidad (porcentaje%)
    df_grouped['Texto'] = df_grouped['Cantidad'].astype(str) + " (" + df_grouped['Porcentaje'].astype(str) + "%)"

    fig = px.bar(df_grouped, x='customer.gender', y='Cantidad', color='Churn', text='Texto',
        barmode='group', color_discrete_sequence=px.colors.qualitative.Dark24,
        labels={'customer.gender': 'Género', 'Churn': 'Cancelación de servicio'}    )

    # Presentación
    fig.update_layout(title='Distribución de Género y Cancelación de Servicio', title_x=0.2,
        xaxis_title='Género', yaxis_title='Cantidad de Clientes', title_font_size=22,
        legend_title_text='¿Canceló el servicio?', bargap=0.2)

    return fig

fig = genero_vs_cancelacion(df_plano)
fig.show()
```

```python
def clientes_adultos_vs_cancelacion(df_plano):
    fig_3 = px.histogram(df_plano, x = 'customer.SeniorCitizen', text_auto=True, color = 'Churn', barmode = 'group',
                color_discrete_sequence= px.colors.qualitative.D3,
                title = 'Clientes Adultos mayores',
                labels= {'customer.SeniorCitizen': '¿Es adulto mayor?', 'Churn': 'Cancelación de servicio'} )
    fig_3.update_layout(yaxis_title='Cantidad de clientes', xaxis_title='Adulto mayor',title_x= 0.4)
    return fig_3
```

```python
fig_3 = clientes_adultos_vs_cancelacion(df_plano)
fig_3.show()
```

### Comparativas de gráficas

```python
# Permanencia vs Tipo de contratación

def permanencia_vs_contrato(df_plano):
    # Copiar para no modificar el DataFrame original
    df = df_plano.copy()

    # Definir etiquetas y orden correcto
    labels = ['0-6 meses', '6-12 meses', '12-24 meses', '24-60 meses', '60+ meses']

    # Crear los cortes con pd.cut()
    df['Rango_permanencia'] = pd.cut(df['customer.tenure'], bins=[0, 6, 12, 24, 60, df['customer.tenure'].max() + 1],  # aseguramos que el último valor caiga en el bin
        labels=labels, right=False, include_lowest=True)

    # Convertir explícitamente a tipo categórico ordenado
    df['Rango_permanencia'] = df['Rango_permanencia'].astype(pd.CategoricalDtype(categories=labels, ordered=True))

    # Crear el gráfico
    fig = px.histogram(df, x='Rango_permanencia', text_auto=True, color='account.Contract', barmode='group',
        color_discrete_sequence=px.colors.qualitative.Bold, title='Permanencia Estimada vs Tipo de Contrato',
        labels={'Rango_permanencia': 'Permanencia estimada', 'account.Contract': 'Tipo de Contratación'},
        category_orders={'Rango_permanencia': labels})  # Reforzamos el orden correcto

    fig.update_layout(yaxis_title='Cantidad de clientes', xaxis_title='Tiempo con el servicio', title_x=0.4)

    return fig

fig_10 = permanencia_vs_contrato(df_plano)
fig_10.show()
```

```python
# Tipo de contrato vs cancelacion
def tipo_contrato_vs_cancelacion(df_plano):
  df_promedio = df_plano.groupby('account.Contract')['customer.tenure'].mean().reset_index()
  fig_11 = px.bar(df_promedio, x='account.Contract', y='customer.tenure', title='Tiempo promedio de permanencia por tipo de contrato',
                  color_discrete_sequence=px.colors.qualitative.Bold)
  fig_11.update_layout(yaxis_title='Tiempo de permanencia', xaxis_title = 'Tipo de Contrato', title_x= 0.4)
  return fig_11

fig_11 = tipo_contrato_vs_cancelacion(df_plano)
fig_11.show()
```

# **Análisis sobre cancelaciones de clientes**
## **Introducción**

Se realizó un analisis con la base de datos proporcionada para comprender el comportamiento de la cancelación de servicios, identificando las oportunidades de mejora para la empresa.

## **Limpieza y tratamiento de los datos**

Se realizó la verificación de la base de datos, para garantizar poder realizar un análisis confiable y así evitar cualquier error que pudiera arrojar la misma.

Se quitaron los espacios vacios, filas vacias y valores nulos, adicionalmente se realizó el cambio de valores binarios por valores de "si" y "no", con la finalidad de tener una mejor facilidad al tratar la información.

## **Análisis Exploratorio de datos**

### *Tipo de contrato*

Se observó una diferencia significativa en el comportamiento de cancelaciones según el tipo de contrato. Presentando que los contratos de largo plazo (dos años) presentan una mayor retención, mientras que los contratos mensuales su tasa de cancelación es mucho mayor.

### **Conclusiones e Insights**

Se hace la recomendación de que al momento de que el cliente haga su contratación, el personal de la empresa promueva lo siguiente:


1.   Contratos de largo plazo (idealmente dos años).
2.   Métodos de pago domiciliados a su tarjeta de crédito.

Con estas prácticas por parte del personal, se podrá contribuir a la mejora de la tasa de retención de clientes y ayuda a evitar cancelaciones a los pocos meses de contratación.


### **Recomendaciones**

Adicionalmente se consideran otras alternetivas a considerar según las posibilidades de la empresa.

Dentro de la alternativas se sugieren las siguientes opciones:



*   Incentivar a los empleados que logren obtener contrataciones de largo plazo.
*   Incentivar al suscriptor a travez de un descuento de pago mensual si realizan la domiciliació de pagos a su tarjeta
*   En caso de que el pago sea mensual, se sugiere que se les de un descuento si llegan a la permanencia minima de un año.
*   Se podría enviar un correo al año para agredecer su permanencia con la empresa y que contenga un cupón para alguna actividad
*   Para el caso de los que contraten por dos años, se le podria proporcionar una cuponera de descuentos o una invitación a un evento exclusivo para los clientes de la empresa.
