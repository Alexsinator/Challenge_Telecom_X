#ðŸ”§ TransformaciÃ³n

```python
df_plano = pd.json_normalize(df.to_dict(orient='records'))
df_plano
```

```python
df_plano.info()
```

```python
for col in df_plano.columns:
  print(f"Valores Ãºnicos en la clumna {col} : {df_plano[col].nunique()}")
  if df_plano[col].unique().size < 10:
    print(df_plano[col].unique())
```

```python
print(df_plano.duplicated().sum()) #Duplicados
```

```python
print(df_plano.isnull().sum()) # Nulos
```

```python
df_plano.apply(lambda x : x.astype(str).str.strip() == '').sum() #Espacios vacios o en blanco
```

```python
df_plano['account.Charges.Total'] = pd.to_numeric(df_plano['account.Charges.Total'], errors='coerce')
df_plano['account.Charges.Total'].dtype #Cambio de tipo a float
```

```python
df_plano = df_plano[df_plano['Churn'].str.strip() != '']
len(df_plano) #Filas vacias
```
